```
# using tensor  used to dervive custome tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```
# other ways

# arange,  it will arrange 0 to 10 with gap 2
print("using arange ->", torch.arange(0,10,2))

# using linspace, it will create  linear space of  9 between o to 10
print("using linspace ->", torch.linspace(0,10,9))

# using eye, It will create the identity metrix of 5 by 5
print("using eye ->", torch.eye(5))

# using full, it will create the sape of 3 by 2 with everyitem is 5
print("using full ->", torch.full((3, 2), 5))
```

Tensor Shapes

```
x = torch.tensor([[1,2,3],[4,5,6]])
x
```

```
x.shape
```

```
# Below is used to create empty tensor with same like above tensor (X) (2 By 3)
torch.empty_like(x)
```

```
# It will create zeros tensor with same shape of x
torch.zeros_like(x)
```

```
torch.ones_like(x)
```

```
# it will not create the tensor of same x shape beacue x is integer values where rand generates float values
torch.rand_like(x)
```

```
# it create the x shape tensor with random values with float beacuse we give the directions
torch.rand_like(x, dtype=torch.float32)
```
