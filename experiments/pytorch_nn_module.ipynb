# Pytorch torch.nn module


```
# create model class
import torch
import torch.nn as nn

class model(nn.modules):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)  #  Input is number of features and output is 1
    self.sigmoid = nn.Sigmoid()
  def forward(self, x):
    out = self.linear(x)
    out = self.sigmoid(out)
    return out
```
